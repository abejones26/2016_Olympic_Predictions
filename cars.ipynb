```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import log
```

```python
cars_data = pd.read_csv("USA_cars_datasets.csv")
cars_data_year = cars_data.loc[cars_data["year"]> 2010,:]
cars_data = cars_data_year[["price", "year"]]
cars_data
```

```python
# N = 500
x = cars_data['year']
y = cars_data['price']
area = np.pi*3
# Plot
plt.scatter(x, y, s=area, marker="o", facecolors="red", edgecolors="black", alpha=0.5)
plt.title('Scatter plot pythonspot.com')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
y = np.log(y)
plt.scatter(x, y, s=area, marker="o", facecolors="red", edgecolors="black", alpha=0.5)
plt.title('Scatter plot pythonspot.com')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
