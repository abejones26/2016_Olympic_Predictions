# Cars Dataset
## Reading Cars csv

```python
ds2 = pd.read_csv("https://abe-southside-final-project.s3.us-east-2.amazonaws.com/USA_cars_datasets.csv")
```

### Features (columns)

```python
ds2 = ds2[["price", "brand", "year", "title_status","mileage", "color", "state"]]
# ds2 = ds2.loc[(ds2["color"] == 'black') | (ds2["color"] == 'white'),:]
ds2
```

## Data Processing

```python
X = pd.get_dummies(ds2[["title_status", "brand", "year", "mileage", "color", "state"]])
y = ds2["price"].values.reshape(-1,1)
print(X.shape, y.shape)
```

```python
X
```

# Scaling and Normalization

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```
